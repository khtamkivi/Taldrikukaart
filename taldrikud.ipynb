Loeme sisse andmestiku. Näeme, et .csv failis sisaldub info külastatud kohtade kohta - kohanimi, laius -ja pikkuskraad, regiooni lühend, milles paik asub ning link kohta tutvustavale Wikipedia lehele.

```python
kohad = pd.read_csv('taldrikud.csv')
print(kohad.shape)
kohad.head(3)
```

Selleks, et eri regioonides asuvaid kohti kaardil paremini eristada, võiks nende tähistamiseks kasutatavaid markereid näiteks erinevalt värvida. Loon selleks lihtsa funktsiooni, millega lisada tabelisse uus tulp, mis sisaldab värvinimetusi.

```python
def select_marker_color(row):
    if row['reg'] == 'E':
        return 'blue'
    elif row['reg'] == 'A':
        return 'green'
    elif row['reg'] == 'F':
        return 'lightred'
    elif row['reg'] == 'P':
        return 'orange'
    elif row['reg'] == 'L':
        return 'pink'
    return 'red'
```

```python
kohad['varv'] = kohad.apply(select_marker_color, axis=1)
kohad.head(3)
```
